````python
custom_stop_words = set(['um', 'uh', 'mm', 'im', 'oh', 'mhm', 'hmm'])
def clean_text(text):
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower()  # Convert to lowercase
    return text

def tokenize_and_lemmatize(text):
    stop_words = set(stopwords.words('english')).union(custom_stop_words)
    tokens = [word for word in word_tokenize(text) if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in tokens]

def preprocess_text(text):
    cleaned_text = clean_text(text)
    lemmatized_tokens = tokenize_and_lemmatize(cleaned_text)
    return ' '.join(lemmatized_tokens)
````

````python
def get_most_frequent_words(texts, top_n=10):
    vectorizer = CountVectorizer()
    bow_matrix = vectorizer.fit_transform(texts)

    # Sum up the counts of each vocabulary word
    word_counts = np.asarray(bow_matrix.sum(axis=0)).flatten()

    # Get the indices that would sort the array in descending order
    sorted_indices = np.argsort(word_counts)[::-1]

    # Get the feature names (words)
    feature_names = vectorizer.get_feature_names_out()

    # Get the top N most frequent words
    most_frequent_words = [(feature_names[idx], word_counts[idx]) for idx in sorted_indices[:top_n]]

    return most_frequent_words
````

````python
# Define a function to generate and display word clouds
def generate_and_display_wordcloud(group_text, group_label):
    # Generate word frequencies
    word_frequencies = WordCloud().process_text(group_text)

    # Filter words with frequency greater than 5
    filtered_words = [(word, freq) for word, freq in word_frequencies.items() if freq > 5]

    # Print the filtered words
    print(f"Words with frequency greater than 5 for {group_label} group:")
    print(filtered_words)
    print("\n")

    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_frequencies), interpolation='bilinear')
    plt.title(f"Word Cloud for {group_label} Group")
    plt.axis('off')
    plt.show()
````

**TF-IDF**

````python
# Function to get the top TF-IDF words for each group
def get_top_tfidf_words(grouped_tokens, top_n=8):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([' '.join(tokens) for tokens in grouped_tokens])
    feature_names = tfidf_vectorizer.get_feature_names_out()
    top_tfidf_words = {}
    for i, tokens in enumerate(grouped_tokens.index):
        tfidf_scores = tfidf_matrix[i].toarray().flatten()
        top_indices = tfidf_scores.argsort()[-top_n:][::-1]
        top_words = [(feature_names[index], tfidf_scores[index]) for index in top_indices]
        top_tfidf_words[tokens] = top_words
    return top_tfidf_words

# Function to print TF-IDF words for random participants in each group
def print_random_tfidf_words(grouped_tokens, group_name):
    top_tfidf_words = get_top_tfidf_words(grouped_tokens, top_n=8)
    print(f"Most Frequent Words (by TF-IDF) for {group_name} Participants:")
    random_participants = random.sample(list(top_tfidf_words.items()), 2)
    for participant_id, words in random_participants:
        print(f"Participant {participant_id}: {words}")

# Function to print the most frequent words for the entire group
def print_most_frequent_words(group_tokens, group_name):
    group_text = ' '.join([str(token) for token in group_tokens])
    word_frequencies = WordCloud().process_text(group_text)
    filtered_words = [(word, freq) for word, freq in word_frequencies.items() if freq > 5]
    print(f"\nMost Frequent Words with frequency greater than 5 for {group_name} group:")
    print(filtered_words)
````

````python
# Function to train Word2Vec and find similar words for a given group
def train_word2vec_and_find_similar_words(group, group_name, word_of_interest):
    # Group data by participant_id and aggregate the 'tokens' column
    grouped_data = group.groupby('Participant_ID')['tokens'].apply(lambda x: ' '.join(x)).reset_index()

    # Tokenize the corpus
    tokenized_corpus = [word_tokenize(doc.lower()) for doc in grouped_data['tokens']]

    # Train the Word2Vec model
    model = Word2Vec(
        sentences=tokenized_corpus,  # The corpus to train the model on
        vector_size=100,             # The size of the word vectors to be learned
        window=5,                    # The size of the window of words to be considered
        min_count=1,                 # Adjusted to 1 to include all words in this small example
        sg=0,                        # 0 for CBOW, 1 for skip-gram
        negative=5,                  # The number of negative samples to use for negative sampling
        ns_exponent=0.75,            # The exponent used to shape the negative sampling distribution
        alpha=0.03,                  # The initial learning rate
        min_alpha=0.0007,            # The minimum learning rate to which the learning rate will be linearly reduced
        epochs=30,                   # The number of epochs (iterations) over the corpus
        workers=4,                   # The number of worker threads to use for training the model
        seed=42,                     # The seed for the random number generator
        max_vocab_size=None          # The maximum vocabulary size (None means no limit)
    )

    # Check if the word is in the vocabulary and retrieve the vector
    if word_of_interest in model.wv:
        vector = model.wv[word_of_interest]
        #print(f"Vector for '{word_of_interest}' in {group_name} group:", vector)

        # Find the most similar words
        similar_words = model.wv.most_similar(word_of_interest)
        print(f"Most similar words to '{word_of_interest}' in {group_name} group:", similar_words)
    else:
        print(f"Word '{word_of_interest}' not found in the vocabulary of {group_name} group.")

````

````python
def get_sentiment_scores(sentences):
    sid = SentimentIntensityAnalyzer()
    sentiment_labels = []
    for sentence in sentences:
      scores = sid.polarity_scores(sentence)
      if scores['compound'] >= 0.05:
          sentiment_labels.append('positive')
      elif scores['compound'] <= -0.05:
          sentiment_labels.append('negative')
      else:
          sentiment_labels.append('neutral')
    return sentiment_labels
````

````python
# Function to get sentiment scores and distribution for a group
def analyze_sentiment_for_group(group, group_name):
    # Initialize VADER sentiment analyzer
    sid = SentimentIntensityAnalyzer()

    # Convert the 'value' column to a list of sentences
    sentences = group['value'].astype(str).tolist()

    # Analyze sentiment for each sentence
    sentiment_labels = get_sentiment_scores(sentences)

    # Count the occurrences of each sentiment
    sentiment_counts = Counter(sentiment_labels)
    total_sentences = len(sentences)

    # Calculate percentage distribution
    sentiment_distribution = {label: count / total_sentences * 100 for label, count in sentiment_counts.items()}

    print(f"Sentiment Distribution for {group_name} group:")
    print(f"Positive: {sentiment_distribution.get('positive', 0):.2f}%")
    print(f"Negative: {sentiment_distribution.get('negative', 0):.2f}%")
    print(f"Neutral: {sentiment_distribution.get('neutral', 0):.2f}%")
    print("-------------------------------------------------------------------")

    # Print sentiment scores for 3 random sentences
    random_sentences = random.sample(sentences, min(3, total_sentences))
    for sentence in random_sentences:
        scores = sid.polarity_scores(sentence)
        print(f"Sentence: {sentence}")
        print(f"Scores: {scores}\n")

# Perform sentiment analysis for each group
for group, group_name in groups_and_names:
    analyze_sentiment_for_group(group, group_name)
````

````python
# Function to preprocess text
def preprocess(tokens):
    # Check if tokens is a string representation of a list
    if isinstance(tokens, str):
        try:
            tokens = ast.literal_eval(tokens)
        except:
            tokens = tokens.split()  # If it's just a string, split it into words

    # Ensure tokens is a list
    if not isinstance(tokens, list):
        return []

    # Convert to lowercase and remove stopwords
    tokens = [token.lower() for token in tokens if token.lower() not in STOPWORDS]

    # Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    return tokens
````

````python
# Preprocess the text for each group
merged_data['processed_tokens'] = merged_data['tokens'].apply(preprocess)

# Function to create and train LDA model for a group
def train_lda_for_group(group_docs, num_topics=10, chunksize=2000, passes=20, iterations=400):
    # Remove any empty lists
    group_docs = [doc for doc in group_docs if doc]

    # Create dictionary and corpus
    dictionary = corpora.Dictionary(group_docs)
    corpus = [dictionary.doc2bow(doc) for doc in group_docs]

    # Train LDA model
    lda_model = LdaMulticore(
        corpus=corpus,
        id2word=dictionary,
        num_topics=num_topics,
        random_state=42,
        chunksize=chunksize,
        passes=passes,
        iterations=iterations,
        eval_every=None
    )

    return lda_model
````

````python
# Prepare data
def prepare_tfidf_data(group):
    vectorizer = TfidfVectorizer(max_features=5000)  # Adjust as needed
    X = vectorizer.fit_transform(group['tokens'])
    feature_names = vectorizer.get_feature_names_out()
    return X.toarray(), feature_names

# Define autoencoder model
def build_autoencoder(input_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(256, activation='relu')(input_layer)
    encoded = Dense(128, activation='relu')(encoded)
    encoded = Dense(64, activation='relu')(encoded)

    decoded = Dense(128, activation='relu')(encoded)
    decoded = Dense(256, activation='relu')(decoded)
    output_layer = Dense(input_dim, activation='sigmoid')(decoded)

    autoencoder = Model(input_layer, output_layer)
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
    return autoencoder

# Train autoencoder for a given group
def train_autoencoder(X):
    input_dim = X.shape[1]
    autoencoder = build_autoencoder(input_dim)
    autoencoder.fit(X, X, epochs=30, batch_size=256, shuffle=True, validation_split=0.2, verbose=1)
    return autoencoder

# Get important words based on reconstruction error
def get_important_words(autoencoder, X, feature_names, top_n=10):
    reconstructed = autoencoder.predict(X)
    reconstruction_error = np.mean(np.abs(X - reconstructed), axis=0)
    important_word_indices = np.argsort(reconstruction_error)[-top_n:]
    important_words = feature_names[important_word_indices]
    return important_words

for group, group_name in groups_and_names:
  group_data, group_feature_names = prepare_tfidf_data(group)

  # Train autoencoder for the group
  autoencoder = train_autoencoder(group_data)

  # Get important words for the group
  important_words = get_important_words(autoencoder, group_data, group_feature_names, top_n=10)

  print(f"Important words for {group_name} group:", important_words)
````

There's a high degree of overlap in important words across all groups. Words like "really," "don't," "yes," "like," "yeah," "know," and "laughter" appear frequently. This suggests that these are common words used in spoken language during interviews, regardless of the group.

The similarity in the top words between the groups could be due to the limited data available, the autoencoder may not be able to capture the more nuanced differences in language patterns between the depressed and not depressed participants.

````python
# Load SpaCy model
nlp = spacy.load('en_core_web_sm')

# Function to extract named entities from a list of texts
def extract_named_entities(texts):
    entities = []
    for text in texts:
        doc = nlp(text)
        for ent in doc.ents:
            entities.append((ent.text, ent.label_))
    return entities

# Function to count entities
def count_entities(entities):
    entity_counter = Counter()
    for entity, label in entities:
        entity_counter[label] += 1
    return entity_counter

# Function to get the most common entities
def most_common_entities(entity_counts, top_n=10):
    return entity_counts.most_common(top_n)

# General function to extract and count named entities for all groups
def extract_and_count_named_entities(groups):
    results = {}

    for group_name, group_data in groups.items():
        entities = extract_named_entities(group_data['tokens'])
        entity_counts = count_entities(entities)
        results[group_name] = {
            "entity_counts": entity_counts,
            "most_common_entities": most_common_entities(entity_counts)
        }

    return results
````

````python
# Define the function to predict depression label for multiple sentences
def predict_sentiments(sentences, tokenizer, model, max_length):
    results = []

    for sentence in sentences:
        # Tokenize and pad the sentence
        sentence_sequence = tokenizer.texts_to_sequences([sentence])
        sentence_padded = pad_sequences(sentence_sequence, maxlen=max_length, padding='post')

        # Predict the label
        predicted_proba = model.predict(sentence_padded)[0][0]

        # Set a threshold to decide the class
        threshold = 0.5
        predicted_label = 1 if predicted_proba >= threshold else 0

        # Map the label to text
        label_mapping = {0: 'Not Depressed', 1: 'Depressed'}
        predicted_label_text = label_mapping[predicted_label]

        # Collect the result
        result = {
            'text': sentence,
            'predicted_label': predicted_label_text,
            'probability_of_being_depressed': predicted_proba * 100,
            'probability_of_not_being_depressed': (1 - predicted_proba) * 100
        }
        results.append(result)

    return results
````
